## Persiapan

### Project Background

Jaya Jaya Institut merupakan salah satu institusi pendidikan perguruan yang telah berdiri sejak tahun 2000. Hingga saat ini ia telah mencetak banyak lulusan dengan reputasi yang sangat baik. Akan tetapi, terdapat banyak juga siswa yang tidak menyelesaikan pendidikannya alias dropout.

Jumlah dropout yang tinggi ini tentunya menjadi salah satu masalah yang besar untuk sebuah institusi pendidikan. Oleh karena itu, Jaya Jaya Institut ingin mendeteksi secepat mungkin siswa yang mungkin akan melakukan dropout sehingga dapat diberi bimbingan khusus.

### Problem Statement

Berikut adalah poin-poin masalah yang dihadapi oleh perusahaan Jaya Jaya Institut:

1. Jaya Jaya Institut mengalami tingkat dropout siswa yang tinggi.
2. Tidak ada sistem deteksi dini untuk mengidentifikasi siswa berisiko *dropout*.
3. Intervensi dilakukan setelah siswa sudah terlambat atau sudah *dropout*.
4. Keterbatasan *monitoring* dan visualisasi data siswa *dropout*.


### Goals

1. Mengembangkan sistem prediksi berbasis data untuk mengidentifikasi siswa Jaya Jaya Institut yang memiliki risiko tinggi untuk *dropout*.
2. Menemukan faktor-faktor utama yang berkontribusi terhadap keputusan mahasiswa untuk *dropout*.
3. Menyediakan rekomendasi strategis bagi manajemen untuk menyusun program bimbingan khusus yang efektif dan tepat sasaran.
4. Memberikan *dashboard* interaktif di Metabase untuk *monitoring* performa mahasiswa secara *real-time*.

### Solution Approach

Untuk mencapai goals yang telah ditetapkan, berikut adalah solusi yang akan diimplementasikan:

1. Menganalisis dataset students' performance untuk memahami karakteristik siswa.
2. Melatih berbagai model untuk memprediksi probabilitas *dropout* seorang mahasiswa, dengan fokus pada variabel target `Status` (khususnya kategori `Dropout`).
3. Menganalisis fitur-fitur yang paling berpengaruh dalam model prediksi untuk memahami pemicu utama *dropout*.
4. Mengevaluasi performa model menggunakan metrik seperti *accuracy*, *precision*, *recall*, dan *F1-score*.
5. Menganalisis *feature importance* untuk mengidentifikasi faktor-faktor utama yang mempengaruhi *dropout*.
6. Membuat *dashboard* interaktif menggunakan Metabase untuk *monitoring* faktor-faktor *dropout*.

### **Deskripsi Dataset**

Dataset ini berasal dari institusi pendidikan tinggi dan diperoleh dari beberapa basis data yang terpisah. Dataset ini berisi data mahasiswa yang terdaftar dalam berbagai program studi sarjana seperti agronomi, desain, pendidikan, keperawatan, jurnalisme, manajemen, pelayanan sosial, dan teknologi. Data mencakup informasi yang diketahui saat pendaftaran mahasiswa (jalur akademik, demografi, dan faktor sosial-ekonomi) serta performa akademik mahasiswa di akhir semester pertama dan kedua. Dataset ini digunakan untuk membangun model klasifikasi guna memprediksi dropout (putus studi) dan kesuksesan akademik mahasiswa.

```python
def comprehensive_data_analysis(df):
    print("=== COMPREHENSIVE DATA ANALYSIS ===")

    # Statistik dasar
    print(f"Dataset Shape: {df.shape}")
    print(f"Memory Usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

    # Analisis target variable
    if 'Status' in df.columns:
        target_dist = df['Status'].value_counts()
        print(f"\nDistribusi Target Variable:")
        for status, count in target_dist.items():
            percentage = (count / len(df)) * 100
            print(f"  {status}: {count} ({percentage:.1f}%)")
    else:
        print("\nKolom 'Status' tidak ditemukan dalam dataset.")

    # Analisis missing values
    print(f"\nMissing Values Analysis:")
    missing_analysis = df.isnull().sum()
    if missing_analysis.sum() == 0:
        print("  Tidak ada missing values dalam dataset.")
    else:
        for col, missing_count in missing_analysis.items():
            if missing_count > 0:
                percentage = (missing_count / len(df)) * 100
                print(f"  {col}: {missing_count} ({percentage:.1f}%)")

    # Analisis duplikat
    print(f"\nDuplicate Rows Analysis:")
    duplicate_count = df.duplicated().sum()
    if duplicate_count == 0:
        print("  Tidak ada data duplikat dalam dataset.")
    else:
        print(f"  Jumlah baris duplikat: {duplicate_count}")

    # Analisis outliers
    print(f"\nOutliers Summary:")
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
    outlier_summary = {}
    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_count = len(outliers)
        outlier_summary[col] = outlier_count

        if outlier_count > 0:
            percentage = (outlier_count / len(df)) * 100
            print(f"  {col}: {outlier_count} ({percentage:.1f}%)")
        else:
            print(f"  {col}: Tidak ada outlier.")

    return outlier_summary
```

Mempertahankan outlier dalam data mahasiswa Jaya Jaya Institut bukanlah sebuah kesalahan, melainkan sebuah strategi yang berorientasi pada inklusivitas, ketelitian, dan akurasi prediksi. Dropout bisa terjadi pada siapa saja, baik dari kelompok umum maupun kelompok minoritas dengan karakteristik ekstrem. Outlier justru memberikan gambaran menyeluruh tentang masalah yang dihadapi oleh institusi. ## Modeling

```python
# Inisiasi Model
model_lr = LogisticRegression(random_state=42)
model_dt = DecisionTreeClassifier(random_state=42)
model_rf = RandomForestClassifier(random_state=42)
model_gb = GradientBoostingClassifier(random_state=42)
model_ab = AdaBoostClassifier(random_state=42)
model_svm = SVC(probability=True, random_state=42)
model_knn = KNeighborsClassifier()
model_gnb = GaussianNB()
model_xgb = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
model_lgbm = LGBMClassifier(random_state=42, verbose=-1)

models = {
    "Logistic Regression": model_lr,
    "Decision Tree": model_dt,
    "Random Forest": model_rf,
    "Gradient Boosting": model_gb,
    "AdaBoost": model_ab,
    "Support Vector Machine": model_svm,
    "K-Nearest Neighbors": model_knn,
    "Gaussian Naive Bayes": model_gnb,
    "XGBoost": model_xgb,
    "LightGBM": model_lgbm
}
```

```python
def evaluate_model(model, X_train, y_train, X_test, y_test, model_name):
    print(f"Evaluating {model_name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]

    # Evaluasi metrik
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    print(f"  Accuracy: {accuracy:.4f}")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall: {recall:.4f}")
    print(f"  F1-Score: {f1:.4f}")
    print(f"  ROC AUC: {roc_auc:.4f}")
    print("-" * 30)

    return {
        'Model': model_name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'ROC AUC': roc_auc,
    }

# Evaluasi semua model
results = []
for name, model in models.items():
    model_results = evaluate_model(model, X_train, y_train, X_test, y_test, name)
    results.append(model_results)

# Buat DataFrame dari hasil
results_df = pd.DataFrame(results)

# Tampilkan hasil dalam tabel
print("\n--- Model Performance Comparison ---")
print(results_df.sort_values(by='Recall', ascending=False))

# Pilih model terbaik berdasarkan Recall
best_model_recall = results_df.sort_values(by='Recall', ascending=False).iloc[0]

print(f"\nModel terbaik berdasarkan Recall adalah: {best_model_recall['Model']}")
print(f"Dengan nilai Recall: {best_model_recall['Recall']:.4f}")
```

Dalam konteks deteksi dini siswa yang berpotensi dropout, metrik recall menjadi pilihan yang sangat penting karena fokus utama dari sistem ini adalah mengidentifikasi sebanyak mungkin siswa yang benar-benar berisiko mengalami dropout.

Memprioritaskan recall berarti mengurangi kemungkinan kesalahan tipe II, yaitu kesalahan saat model gagal mendeteksi siswa yang sebenarnya akan dropout. Kesalahan seperti ini bisa berdampak serius karena siswa yang membutuhkan intervensi justru tidak terdeteksi dan tidak mendapatkan bantuan tepat waktu. Dalam dunia pendidikan, kehilangan satu siswa karena tidak terdeteksi risikonya lebih merugikan daripada memberikan perhatian tambahan kepada siswa yang sebenarnya tidak berisiko tinggi (false positive).

Dengan demikian, menggunakan recall sebagai metrik utama sejalan dengan tujuan sosial dan praktis dari deteksi dini dropout, yakni mencegah sebanyak mungkin kasus putus sekolah melalui identifikasi yang sensitif terhadap siswa-siswa yang rentan.

```python
# Feature Importance

# Melatih model XGBoost dengan data pelatihan
model_xgb.fit(X_train, y_train)

# skor penting setiap fitur dari model
importances = model_xgb.feature_importances_

# DataFrame untuk menyimpan nama fitur dan tingkat kepentingannya
importance_df = pd.DataFrame({
    'Fitur': X_train.columns,
    'Skor_Kepentingan': importances
})

# Mengurutkan fitur berdasarkan skor kepentingan secara menurun
importance_df = importance_df.sort_values(by='Skor_Kepentingan', ascending=False)

# visualisasi batang horizontal
plt.figure(figsize=(15, 8))
ax = sns.barplot(x='Skor_Kepentingan', y='Fitur', data=importance_df, palette='viridis')

# label
for bar in ax.patches:
    bar_width = bar.get_width()
    ax.annotate(f'{bar_width:.3f}',
                xy=(bar_width + 0.005, bar.get_y() + bar.get_height() / 2),
                va='center', ha='left',
                fontsize=8, color='black')

# judul dan label sumbu
plt.title('Tingkat Kepentingan Fitur - XGBoost', fontsize=14)
plt.xlabel('Skor Kepentingan', fontsize=12)
plt.ylabel('Nama Fitur', fontsize=12)
plt.tight_layout()
plt.show()
```

Berdasarkan visualisasi tingkat kepentingan fitur pada model XGBoost, terlihat bahwa prediktor paling signifikan terhadap kemungkinan dropout siswa adalah faktor-faktor yang berkaitan dengan performa akademik dan kondisi keuangan. Fitur Curricular_units_2nd_sem_approved dan Tuition_fees_up_to_date menempati peringkat teratas dengan skor kepentingan yang sangat tinggi, yaitu masing-masing sebesar 0.256 dan 0.212. Hal ini menunjukkan bahwa jumlah mata kuliah yang diselesaikan pada semester kedua serta keteraturan pembayaran biaya kuliah memiliki pengaruh besar terhadap keberlanjutan studi siswa. Selain itu, keikutsertaan dalam beasiswa (Scholarship_holder) dan tingkat partisipasi akademik pada semester pertama dan kedua juga termasuk dalam fitur yang sangat berkontribusi. Di sisi lain, fitur-fitur yang berkaitan dengan latar belakang demografis, seperti jenis kelamin, kebangsaan, pekerjaan orang tua, serta status pernikahan, memiliki skor kepentingan yang jauh lebih rendah. Temuan ini menegaskan bahwa dalam konteks deteksi dini siswa dropout, indikator akademik awal dan aspek finansial merupakan komponen kunci yang perlu menjadi fokus utama intervensi, sementara faktor sosial ekonomi dan demografis cenderung tidak terlalu menentukan.
